 ## Data Visualization using Matplotlib and Seaborn
 
With the help of sample dataframe, I have explored here various data visualization plots such as histplot, bar plot,box plot, KDE plot, stacked bar chart, scatter plot, heatmap using Mstplotlib and Seaborn.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_csv('df1',index_col=0)
df1.head()
```

```python
df2=pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist',bins=30)
```

```python
plt.style.use('ggplot')
df1['A'].plot.hist(bins=30)
```

```python
df1['A'].hist(bins=30)
```

```python
plt.style.use('bmh')
df1['A'].hist(bins=30)
```

```python
plt.style.use('dark_background')
df1['A'].hist(bins=30)
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist(bins=30)
```

```python
plt.style.use('ggplot')
df2.plot.area(alpha=1)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

```python
df2.plot.box() # Can also pass a by= argument for groupby
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
